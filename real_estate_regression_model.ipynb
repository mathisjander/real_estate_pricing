```python
import pandas as pd

df = pd.read_csv('real_estate_listings_clean.csv')
df
```

```python
# drop url col

df = df.drop(['url', 'energy', 'heating', 'fee'], axis=1)
df
```

```python
y = df['price']
X = df.iloc[:,1:-1]
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split


# Perform train-test split (e.g., 70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor

# Create a sample dataset
X, y = make_regression(n_samples=1000, n_features=10, noise=0.1, random_state=42)

# Define the models
models = {
    'linear_regression': {
        'model': LinearRegression(),
        'params': {
            'fit_intercept': [True, False],
            'n_jobs': [-1]
        }
    },
    'ridge_regression': {
        'model': Ridge(),
        'params': {
            'alpha': [1, 5, 10, 50, 100],
            'fit_intercept': [True, False],
            'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
        }
    },
    'lasso_regression': {
        'model': Lasso(),
        'params': {
            'alpha': [0.1, 0.5, 1, 5, 10],
            'fit_intercept': [True, False],
            'max_iter': [1000, 2000, 5000],
            'selection': ['cyclic', 'random']
        }
    },
    'xgboost': {
        'model': XGBRegressor(),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7, 10],
            'subsample': [0.5, 0.7, 1],
            'colsample_bytree': [0.5, 0.7, 1]
        }
    },
    'random_forest': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [10, 50, 100, 200],
            'criterion': ['squared_error', 'absolute_error'],
            'max_depth': [None, 5, 10, 15, 20],
            'min_samples_split': [2, 5, 10, 20]
        }
    }
}

# Perform RandomizedSearchCV for each model and store the results
results = []

for model_name, model_info in models.items():
    random_search = RandomizedSearchCV(
        estimator=model_info['model'],
        param_distributions=model_info['params'],
        n_iter=10,
        cv=5,
        scoring='neg_mean_squared_error',
        n_jobs=-1,
        random_state=42
    )
    random_search.fit(X_train, y_train)
    results.append({
        'model': model_name,
        'best_params': random_search.best_params_,
        'best_score': -random_search.best_score_
    })

# Display the results as a pandas DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# create a grid search for the x
from sklearn.model_selection import GridSearchCV

xgb_params = {'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7, 10],
            'subsample': [0.5, 0.7, 1],
            'colsample_bytree': [0.5, 0.7, 1]}

grid_search = GridSearchCV(
        estimator=XGBRegressor(),
        param_grid=xgb_params,
        cv=5,
        scoring='neg_mean_squared_error',
        n_jobs=-1
    )
grid_search.fit(X_train, y_train)
-grid_search.best_score_
```

```python
from sklearn.metrics import mean_squared_error

final_model = grid_search.best_estimator_

y_pred = final_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f'Root Mean Squared Error: {mse**0.5}')
```

```python
# pickle final model
import pickle

pickle.dump(final_model, open('app/real_estate_model.pkl', 'wb'))

# pickle the scaler

pickle.dump(scaler, open('app/input_scaler.pkl', 'wb'))
```

```python
# create shap explainer for prediction
import shap
import matplotlib.pyplot as plt

explainer = shap.TreeExplainer(final_model)
shap_values = explainer.shap_values(X_test)

shap.initjs()

shap.force_plot(
    base_value=explainer.expected_value,
    shap_values=shap_values[0,:],
    features=X_test[0,:],
    feature_names=df.columns[:-1]
)

```
